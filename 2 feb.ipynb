

For loop: A for loop is a control flow statement that executes code repeatedly for a particular number of iterations. In this control flow statement, the keyword used is for. The for loop is used when the number of iterations is already known.

```python
#Answer:-1
l=[1,2,3,4,5]

for i in range(len(l)):
    print(l[i])
```

Q2.Write a python program to print the sum and product of the first 10 natural numbers using for and while loop.

```python
#Answer :-2

#using for loop

sum1=0
product=1
for i in range(1,11):
    sum1=sum1+i
    product=product*i
    
print("Sum of first 10 natural number is:",sum1)
print("Product of first 10 natural number is:",product)
```

```python
#Answer :-2

#using while loop

sum1=0
product=1
i=1
while i<=10:
    sum1=sum1+i
    product=product*i
    i=i+1
    
print("Sum of first 10 natural number is :",sum1)
print("Product of first 10 natural number is :",product)
```

```python
#Answer :-3

units=int(input("Enter the unit of electricity consumed in a month:-"))

if units<=100:
    bill=units*4.5
elif units<=200:
    bill=(100*4.5)+(units-100)*6
elif units<=300:
    bill=(100*4.5)+(100*6)+(units-200)*10
else:
    bill=(100*4.5)+(100*6)+(100*10)+(units-300)*20
    
print("Total Electricity bill is :-",bill)
```

```python
#Answer:-4

#using for loop

l1=[]
for i in range(1,101):
    cube=i**3
    if cube%4==0 or cube%5==0 :
        l1.append(i)
        
print("List of numbers whose cube is divisible by 4 or 5:",l1)
```

```python
#using while loop

l2=[]
i=1
while i<=100:
    cube=i**3
    if cube%4==0 or cube%5==0 :
        l2.append(i)
    i=i+1
        
print("List of numbers whose cube is divisible by 4 or 5:",l2)
```
